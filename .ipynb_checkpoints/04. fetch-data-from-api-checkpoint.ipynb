```python
for i in range(1,525):
    response  = req.get('https://api.themoviedb.org/3/movie/top_rated?api_key=e86d2714841ecb3d4f2f1ff7ff087447&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()["results"])[['id','overview','popularity','release_date','title','vote_average','vote_count']]
    df = df.append(temp_df, ignore_index=True)
```
